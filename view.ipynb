```python
# Import library
import os
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
# Menampilkan dataset Bonn Subset Z

# Define the folder and file naming scheme
folder_path = 'BONN_SVM-KSRC\\Bonn_Raw_Dataset\\A_Z'
file_prefix = 'Z'
file_extension = '.txt'
file_count = 100

# Initialize an empty dictionary to hold the data for each file
data = {}

# Loop through each file and read its contents
for i in range(1, file_count + 1):
    file_name = f'{file_prefix}{str(i).zfill(3)}{file_extension}'
    file_path = os.path.join(folder_path, file_name)
    
    # Read the file contents
    with open(file_path, 'r') as file:
        # Read all lines and strip any leading/trailing whitespace
        lines = [line.strip() for line in file.readlines()]
        # Store the lines in the dictionary
        data[f'Segment_{i}'] = lines

# Convert the dictionary to a DataFrame
df = pd.DataFrame.from_dict(data, orient='index').transpose()

# Show the DataFrame
print('Bonn Subset Z')
print(df)
```

```python
# Menampilkan dataset NSC-ND Subset X

# Define the folder and file naming scheme
folder_path = 'NSC-ND_SVM-KSRC\\NewDelhi_Raw_Dataset\\X'
file_prefix = 'preictal'
file_extension = '.mat'
file_count = 50

# Initialize an empty dictionary to hold the data for each file
data = {}

# Loop through each file and read its contents
for i in range(1, file_count + 1):
    file_name = f'{file_prefix}{i}{file_extension}'
    file_path = os.path.join(folder_path, file_name)
    
    # Read the .mat file contents
    mat_data = loadmat(file_path)
    
    # Extract the relevant data from the .mat file
    lines = mat_data[file_prefix]
    # Convert the numpy array to a list
    lines_list = lines.flatten().tolist()
    # Store the lines in the dictionary
    data[f'Segment_{i}'] = lines_list

# Convert the dictionary to a DataFrame
df = pd.DataFrame.from_dict(data, orient='index').transpose()

# Print the entire DataFrame
print('NSC-ND Subset X')
print(df)
```
